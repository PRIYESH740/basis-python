```python
def printkfreqwords(s,k):
    words=s.split()
    d={}
    for w in words:
        d[w]=d.get(w,0)+ 1
        
    for w in d:
        if d[w]==k:
            print(w)
        
    
    
```

```python
def sumUnique(l):
    s=set()
    for i in l:
        s.add(i)
    sum=0
    for i in s:
        sum=sum+i
    return sum
```

```python
sumUnique([1,2,3,4,4,3,4,2,1,5,5])
```
