```python
while True:
  n =int(input())
  if n==1:
    x=int(input())
    y=int(input())
    print(x+y)

  elif n==2:
    x=int(input())
    y=int(input())
    print(x-y)
  elif n==3:
    x=int(input())
    y=int(input())
    print(x*y)
  elif n==4:
    x=int(input())
    y=int(input())
    print(int(x/y))
  elif n==5:
    x=int(input())
    y=int(input())
    print(x%y)
  elif n==6:
    quit()

  else:
    print("Invalid Operation")
```

### Reverse of a number

```python
num=1234
reversed_num =0
while  num!=0:
    digit =num%10
    reversed_num =reversed_num*10 + digit
    num //=10

print("Reversed Number: "+str(reversed_num))
```
