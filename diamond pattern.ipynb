```python
n=int(input())
i=1
while i<=n:
    space=2
    while space>=i and i<=2:
        print(" ",end='')
        space -=1
        
    j=1
    while j<=i and i<3:
        print("*",end='')
        j +=1
        
        
    star=i-1
    while star>=1 and i<=3:
            print("*",end='')
            star -=1
            
    space1=i-3
    while space1>=1:
        print(" ",end='')
        space1 -=1
        
    star1=1
    while star1<=n-i+1 and i>=3:
        print("*",end='')
        star1 +=1
        
        
    star2=i-3
    while star2<=n-i and i>3:
        print("*",end='')
        star2 +=1
        
    print()
    i +=1
```

```python
n = int(input())
cols = 2*n + 1
i = 1

while i <= n:
    j = 1
    while j <= cols:
        if(j == i or j == (cols // 2 + 1) or j ==cols- i +1):
            print('*',end = '')
            
        else:
            print('0',end = '')

        j = j + 1
        
    print()
    i = i + 1
```

```python
n=int(input())

for i in range (1,n+1,1):
    for j in range(i-1):
        print(" ",end='')
        
    for k in range(i,n+1,1):
        print(k,end='')
    print()
    

# down pattern 
for i in range (1,n,1):
    for j in range(n-i-1):
        print(" ",end='')
        
    for k in range(n-i,n+1,1):
        print(k,end='')
    print()
        
```
