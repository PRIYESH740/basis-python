### nth  fibonacci number

```python
def fib_num(n):
    a=0
    b=1
    if n<=0:
        print("Fibonacci can't be possible ")
    
    elif n==1:
        return 0
    #elif n==2:
        #return 1
    else:
        for i in range(2,n+1):
            c=a+b
            a=b
            b=c
        return b
    
n=int(input())
print(fib_num(n))
```

```python
def checkMember(n):
    a=0
    b=1
    if a>n:
        return False
    elif a==n:
        return True
    else:
        while b<n:
            c=a+b
            a=b
            b=c
            if b==n:
                return True
            else:
                return False
    
n=int(input())
if(checkMember(n)):
    print("true")
else:
    print("false")
```
