### sum of even number

```python
n = int(input())
i=1
sum=0
while i<=n:
    if i%2==0:
         sum=sum+i
         i=i+1
    else:
        i=i+1

print(sum)
```

```python
num=101
temp=num
reversed_num =0
while  num!=0:
    digit =num%10
    reversed_num =reversed_num*10 + digit
    num //=10
    
if reversed_num==temp:
    print('true')
    
else:
    print('false')
```

```python
def checkPalindrome(num):
    temp=num
    reversed_num =0
    while  num!=0:
         digit =num%10
         reversed_num =reversed_num*10 + digit
         num //=10
    
    if reversed_num==temp:
         return True
    
    else:
         return False

num = int(input())
isPalindrome = checkPalindrome(num)
if(isPalindrome):
    print('true')
else:
    print('false')
```

### Sum of even and odd number in given number 1024

```python
num=int(input())
sum1=0
sum2=0

while num!=0:
    digit= num%10
    if (digit%2)==0:
        sum1=sum1+digit
        num //=10
    else:
        sum2=sum2+digit
        num //=10
        
print(sum1,"",sum2)
```

### nth fibonacci number


```python
num=int(input())

a=0
b=1
if(num==1):
    print(a)
while num!=0:
        c=a+b
        a=b
        b=c
        num -=1
    
print(c)
```
