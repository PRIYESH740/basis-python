```python
def sum(a,b,c=0,d=0,e=0):
    return a+b+c+d+e
```

```python
sum(1,2,3,4,5)
```

```python
def sum2(a,b,*more):
    print(a)
    print(b)
    print(type(more))
    ans=a+b
    for i in more:
        ans=ans+i
    return ans
```

```python
sum2(2,3,4,5)
```

```python
def sum_diff(a,b):
    return a+b, a-b
```

```python
c=sum_diff(4,1)
c
```

```python
d,e=sum_diff(4,1)
print(d)
print(e)
```
