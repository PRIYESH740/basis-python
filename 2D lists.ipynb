```python
def merge(a1, a2, arr):
    i = 0
    j = 0
    k = 0
    
    while i < len(a1) and j <len(a2):
        if a1[i] < a2[j]:
            arr[k] = a1[i]
            k = k + 1
            i = i + 1
        else:
            arr[k] = a2[j]
            k = k + 1
            j = j + 1
    
    while i < len(a1):
        arr[k] = a1[i]
        k = k + 1
        i = i + 1
    
    while j < len(a2):
        arr[k] = a2[j]
        k = k + 1
        j = j + 1
    
def MergeSort(arr,l,r):
    if len(arr) == 0 or len(arr) == 1:
        return 
    
    mid = len(arr)//2
    
    a1 = arr[:mid]
    a2 = arr[mid:]
    
    MergeSort(a1,l,mid-1)
    MergeSort(a2,mid,r)
    
    merge(a1, a2, arr)
    
    
    
    
arr = [5,2,45,12,1,6,52,41,11]
MergeSort(arr,0,8)
print(arr)
```
