```python
def isPalindrome(str):
    if start_index >= end_index:
        return True
    
    if str[0]!=str[len(str)-1]:
        return False
    
    reverse=str[1:len(str)-1]
    ans=isPalindrome(reverse)
    
    if ans:
        return True
    else:
        return False

str="abccba"
ans=isPalindrome(str)
print(ans)
if ans:
    print("Palindrome")
    
else :
    print("not Palindrome")    
```

```python
def checkAB(str):
    if len(str) == 0:
        return True
    if len(str) == 1:
        if str == 'a':
            return True
        else:
            return False
    if str[0] == 'a':
        return checkAB(str[1:])
    elif str[0] == 'b':
        if str[1] == 'b':
            return checkAB(str[2:])
        else:
            return False
    else:
        return False
    
str= input()
ans=checkAB(str)
if ans:
    print('true')
    
else:
    print('false')
    
```
