```python
def pushZerosAtEnd(arr, n) :
    for i in range(n-1):
        for j in range(i,n):
            if arr[j]!=0:
                arr[i],arr[j]=arr[j],arr[i]
                break
#arr=[2,0,0,1,3,0,0] 
#arr=[0,3,0,2,0]
arr=[9,0,0,8,2]
n=len(arr)
pushZerosAtEnd(arr, n)
print(arr)
```

```python
# function to rotate array by d elements using temp array
def rotateArray(arr, n, d):
    temp = []
    i = 0
    while (i < d):
        temp.append(arr[i])
        i = i + 1
    i = 0
    while (d < n):
        arr[i] = arr[d]
        i = i + 1
        d = d + 1
    arr[:] = arr[: i] + temp
    

d =int(input())
arr=[1,2,3,4,5,6,7]
#arr=[1,2,3,4]

n=len(arr)
rotateArray(arr, n,d) 
print(arr)
```

```python
def sort(arr,n):
    arr1=[]
    temp=arr.count(0)
    temp1=arr.count(1)
    temp2=arr.count(2)
    for i in range(temp):
        arr1.append(0)

    for i in range(temp1):
        arr1.append(1)

    for i in range(temp2):
        arr1.append(2)

    for i in range(n):
        arr[i]=arr1[i]

arr=[0,1,2,0,2,0,1]
n = len(arr)
sort(arr,n)
print(arr)
```

```python
def sum_arrays(array1,n, array2,m):
    # Ensure both arrays have the same length
    if n!=m:
        raise ValueError("Arrays must have the same length")

    # Calculate the sum of each pair
    sum_array = [x + y for x, y in zip(array1, array2)]
    
    return sum_array

# Example usage
array1 = [1, 2, 3]
array2 = [4, 5, 6]
result = sum_arrays(array1, array2)
print(result)  # Output: [5, 7, 9]
```
