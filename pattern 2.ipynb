### inverted pattern

```python
n=int(input())
i=1

while i<=n:
    j=1
    while j<=n-i+1:
        print("*",end='')
        
        j +=1
    print()
    i +=1
```

```python
n=int(input())
i=1
k=n
while i<=n:
    j=1
    while j<=k:
        print(k,end='')
        
        j +=1
    print()
    k -=1
    i +=1
```

```python
n=int(input())
i=1
while i<=n:
    j=1
    k=1
    while k>=n-i:
        print(" ",end='')
        k +=1
    while j<=i:
        print("*",end='')
        
        j +=1
    print()
    i +=1
```

```python
n=int(input())
i=1
while i<=n:
    space=1
    while space<=n-i:
        print(" ",end='')
        space +=1
    # upward pyramid 
    num=1
    while num<=i:
        print(num,end='')
        num +=1
        
    num=i-1
    while num>=1:
        print(num,end='')
        num -=1
    print()
    i +=1
        
    # downward pyramid 
x=1   
while x<n:
    space2=1
    while space2<=x:
        print(" ",end='')
        space2 +=1
        
    j=1
    while j<=n-x:
        print(j,end='')
        j +=1
        
    num1=n-x-1
    while num1>=1:
        print(num1,end='')
        num1 -=1
        
    print()
    x +=1
```
