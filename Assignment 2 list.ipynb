```python
def is_permutation(a, b):
    if len(a) != len(b):
        return False

    char_count = [0] * 256
    for c in a:
        char_count[ord(c)] += 1

    for c in b:
        char_count[ord(c)] -= 1
        if char_count[ord(c)] < 0:
            return False

    return True
```

```python
def reverse_words(string):
    list1 = string.split()
    list2 = []
    for e in list1:
        e = e[::-1]
        list2.append(e)
    return ' '.join(list2)

string="Welcome to coding ninja"
print(reverse_words(string))
```

```python
def getCompressedString(input) :
    i = 0
    x = ''
    while(i < len(input)):
        j = i+1
        c = 1
        while j < len(input) and (input[i] == input[j]):
            j += 1
            c += 1
        if c == 1:
            x += input[i]
        else:
            x += input[i]+str(c)
        i = j
    return x


input = input()
print(getCompressedString(input))
```

```python
def highest_occurrence(string):
    char_count = {}
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    highest_count = 0
    highest_char = None
    for char, count in char_count.items():
        if count > highest_count:
            highest_count = count
            highest_char = char

    return highest_char

# Example usage
text = "abcdeapapqarr"
result = highest_occurrence(text)
print("The highest occurring character is:", result)
```

```python
def equilibrium(arr):
 
    
    total_sum = sum(arr)
    leftsum = 0
    for i, num in enumerate(arr):
         
        total_sum -= num
         
        if leftsum == total_sum:
            return i
        leftsum += num
    
    return -1
     
# Driver code
arr = [1,-1,4]
print ('First equilibrium index is ',
       equilibrium(arr))
```

```python
def find3Numbers(arr, n, num):
    arr.sort()
    count = 0
    for i in range(n - 1):
        d = {}
        curr_sum = num - arr[i]
        for j in range(i + 1, n):
            if curr_sum - arr[j] in d:
                count += 1
            d[arr[j]] = True
    return count

#arr = [1,2,3,4,5,6,7]
#arr=[2,-5, 8,-6,0,5,10,11,-3]

arr=[6,1,6,5,3,2,5,0,5,6,0 ]
#arr=[]
n=len(arr)
num = 5
print(find3Numbers(arr,n, num))
```
