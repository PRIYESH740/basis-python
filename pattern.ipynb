```python
n=int(input())
i=1
while i<=n:
    j=1
    #print the first triangluar pattern 
    while j<=n-i+1:
        print(j,end='')
        j +=1
    
    #print second triangluar pattern    
    start=i-1
    while start:
        print("**",end='')
        start -=1
   
    #print the third triangular pattern    
    last =1
    while last<=n-i+1:
        print(n-last-i+2,end='')
        last +=1
        
    print()
    i +=1
```

```python
n=int(input())
i=1
while i<=n:
    space=2
    while space<=i and i<=2:
        print(" ",end='')
        space -=1
        
    j=1
    while j<=i and i<=3:
        print("*",end='')
        j +=1
        
        
    star=i-1
    while star>=1 and i<=3:
            print("*",end='')
            star -=1
            
    space1=i-3
    while space1>=1:
        print(" ",end='')
        space1 -=1
        
    star1=1
    while star1<=n-i+1 and i>3:
        print("*",end='')
        star1 +=1
        
        
    star2=i-3
    while star2<=n-i and i>3:
        print("*",end='')
        star2 +=1
        
    print()
    i +=1
```
