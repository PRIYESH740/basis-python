```python
def rotate(arr, n, d):
    if n==0:
        return 
    arr=arr[d:]+arr[:d]
    return arr
    

#arr=[1,2,3,4,5,6,7]
#arr=[1,2,3,4]
arr=[1,3,6,11,12,17]
n=len(arr)
d = 4
ans=rotate(arr,n, d)
print(ans)
```

```python
# List slicing approch to rotate the array
def rotateArray(arr,n,d):
    arr[:]=arr[d:n]+arr[0:d]
    return arr

arr = [1, 2, 3, 4, 5, 6]
n=len(arr)

print("Rotated list is")
print(rotateArray(arr,n,2)) 
```

```python
def find3Numbers(arr, n, num):
    count = 0
    arr.sort()
    for i in range(n - 2):
        l = i + 1
        r = n - 1
        curr_sum = num - arr[i]
        while l < r:
            if arr[l] + arr[r] == curr_sum:
                count += 1
                l += 1
                r -= 1
            elif arr[l] + arr[r] < curr_sum:
                l += 1
            else:
                r -= 1
    return count

#arr = [1,2,3,4,5,6,7]
#arr=[2,-5, 8,-6,0,5,10,11,-3]
arr=[6,1,6,5,3,2,5,0,5,6,0 ]
#arr=[]
n=len(arr)
num = 5
print(find3Numbers(arr,n, num))
```
